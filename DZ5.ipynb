Задача 1. Вы провели эксперимент c упрощением формы заказа в магазине Утконос и получили результаты по метрике конверсий в покупку. Выберите метод оценки и оцените есть ли стат.значимые различия между конверсиями в двух группах при alpha = 5%. Дайте краткие рекомендации команде .Результаты: 1. Число юзеров в группах , которые заходили на сайт в период эксперимента: n1 = 15550 и n2 = 15550 . 2.  Число юзеров в группах , которые совершили хотя бы одну покупку за период эксперимента: n1 = 164 и n2 = 228 3.  Конверсии : conv1 = 1.05% conv2 = 1.47% .

```python
from math import sqrt


p1 = 164 / 15550 # конверсия в первой группе

p2 = 228 / 15550 # конверсия во второй группе

n1 = 164 # число юзеров совершивших покупку в 1 группе

n2 = 228 # число юзеров совершивших покупку во 2 группе

N1 = 15550 # общее число юзеров в 1 первой группе

N2 = 15550 # общее число юзеров во 2 первой группе

p =  (n1 * p1 + n2 * p2) / (n1 + n2)

Z = (p1 - p2) / sqrt(p * (1 - p) * (1/N1 + 1/N2))
Z
```

Вывод: Критическое значение Z-критерия при α=5% и двухсторонней альтернативе равно 1.96. Расчетное значение Z меньше критического значения,   мы отвергаем нулевую гипотезу и  рекомендуем использовать упрощенную форму заказа.
